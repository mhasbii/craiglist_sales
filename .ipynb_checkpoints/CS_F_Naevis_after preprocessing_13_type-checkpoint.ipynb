```python
# Mencari jumlah missing value
def cek_null(df):
    col_na = df.isnull().sum().sort_values(ascending=False)
    percent = col_na / len(df)
    
    missing_data = pd.concat([col_na, percent], axis=1, keys=['Total', 'Percent'])
    print(missing_data[missing_data['Total'] > 0])

print("Jumlah missing values untuk Vehicle data: \n")
print(cek_null(vehicles))
```

Terdapat sangat banyak missing value pada data vehicles 

```python
# Mencari duplikasi data
sum_duplicate = vehicles.duplicated().sum()
print("Jumlah data yang redundant: ", sum_duplicate)
```

```python
# Menghitung outlier
Q1 = vehicles.quantile(0.25)
Q3 = vehicles.quantile(0.75)
IQR = Q3 - Q1
print("Jumlah data yang merupakan outlier")
((vehicles < (Q1 - 1.5 * IQR)) | (vehicles > (Q3 + 1.5 * IQR))).sum()
```

### Jenis Kendaraan Paling Populer

```python
# Manufacture
grouped_manu = vehicles.groupby(["state", "manufacturer"]).size().reset_index(name='counts')
max_val_manu = grouped_manu.groupby(['state'])['counts'].transform(max) == grouped_manu['counts']

grouped_manu[max_val_manu]
```

```python
# Model
grouped_model = vehicles.groupby(["state", "model"]).size().reset_index(name='counts')
max_val_model = grouped_model.groupby(['state'])['counts'].transform(max) == grouped_model['counts']

grouped_model[max_val_model]
```

```python
# Fuel
grouped_fuel = vehicles.groupby(["state", "fuel"]).size().reset_index(name='counts')
max_val_fuel = grouped_fuel.groupby(['state'])['counts'].transform(max) == grouped_fuel['counts']

grouped_fuel[max_val_fuel]
```

```python
# Transmission
grouped_trans = vehicles.groupby(["state", "transmission"]).size().reset_index(name='counts')
max_val_trans = grouped_trans.groupby(['state'])['counts'].transform(max) == grouped_trans['counts']

grouped_trans[max_val_trans]
```

```python
# Drive 
grouped_vehicles = vehicles.groupby(["state", "drive"]).size().reset_index(name='counts')
max_val = grouped_vehicles.groupby(['state'])['counts'].transform(max) == grouped_vehicles['counts']

grouped_vehicles[max_val]
```

```python
# Size
grouped_size = vehicles.groupby(["state", "size"]).size().reset_index(name='counts')
max_val_size = grouped_size.groupby(['state'])['counts'].transform(max) == grouped_size['counts']

grouped_size[max_val_size]
```

```python
# Type
grouped_type = vehicles.groupby(["state", "type"]).size().reset_index(name='counts')
max_val_type = grouped_type.groupby(['state'])['counts'].transform(max) == grouped_type['counts']

grouped_type[max_val_type]
```

```python
# Paint Color
grouped_paint = vehicles.groupby(["state", "paint_color"]).size().reset_index(name='counts')
max_val_paint = grouped_paint.groupby(['state'])['counts'].transform(max) == grouped_paint['counts']

grouped_paint[max_val_paint]
```

# Data Pre-Processing

```python
# Copy Data
data_harga = vehicles.copy()
data_harga.head()
```

#### Handle Missing Value

```python
# Missing Values Function
def cek_missvalue(df):
  col_mv = df.isnull().sum().sort_values(ascending=False)
  percent = (col_mv / len(df))*100

  missing_data = pd.concat([col_mv, percent], axis=1, keys=["Total Missing Value", "Percent"])
  print(missing_data[missing_data['Total Missing Value'] > 0])
```

```python
# Check Missing Value
cek_missvalue(data_harga)
```

```python
data_harga.drop(['size'], axis=1, inplace=True)
data_harga.drop(['cylinders'], axis=1, inplace=True)
data_harga.drop(['paint_color'], axis=1, inplace=True)
data_harga.drop(['drive'], axis=1, inplace=True)
data_harga.drop(['VIN'], axis=1, inplace=True)
data_harga.drop(['lat'], axis=1, inplace=True)
data_harga.drop(['long'], axis=1, inplace=True)
data_harga.drop(['description'], axis=1, inplace=True)
data_harga.drop(['posting_date'], axis=1, inplace=True)
data_harga.drop(['image_url'], axis=1, inplace=True)
data_harga.drop(['url'], axis=1, inplace=True)
data_harga.drop(['id'], axis=1, inplace=True)
data_harga.drop(['region_url'], axis=1, inplace=True)
data_harga.drop(['model'], axis=1, inplace=True)
# data_harga.drop(['transmission'], axis=1, inplace=True)
```

```python
# Handle Missing Value
data_harga.dropna(inplace=True)
```

```python
cek_missvalue(data_harga)
```

```python
# Cek duplikasi data
dup = data_harga[data_harga.duplicated()]
print("Data yang memiliki duplikasi ada sebanyak:\n", dup.count())
```

```python
data_harga
```

```python
# Outlier handler

# def outlierChecker(data):
q1 = data_harga.quantile(0.25)
q3 = data_harga.quantile(0.75)
iqr = q3 - q1
    
print("Jumlah data sebelum: \n", data_harga.count())
data_harga = data_harga[~((data_harga < (q1 - 1.5 * iqr)) |(data_harga > (q3 + 1.5 * iqr))).any(axis=1)]
print("Jumlah data setelah: \n", data_harga.count())
```

```python
# Label Encoder
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
data_clean['region'] = label.fit_transform(data_clean['region'])
data_clean['manufacturer'] = label.fit_transform(data_clean['manufacturer'])
data_clean['condition'] = label.fit_transform(data_clean['condition'])
data_clean['fuel'] = label.fit_transform(data_clean['fuel'])
data_clean['title_status'] = label.fit_transform(data_clean['title_status'])
data_clean['transmission'] = label.fit_transform(data_clean['transmission'])
# clean_data['type'] = label.fit_transform(clean_data['type'])
data_clean['state'] = label.fit_transform(data_clean['state'])
```

### Van

```python
data_van = data_clean.loc[(data_clean['type'] == 'van')]
data_van
```

```python
sns.histplot(data=data_van, x="price", kde=True)
```

```python
q1_price = data_van['price'].quantile(0.25)
q3_price = data_van['price'].quantile(0.75)
iqr = q3_price - q1_price
lower_price = q1_price - (1.5*iqr)
upper_price = q3_price + (1.5*iqr)

data_van = data_van.loc[(data_van['price'] <= upper_price) & (data_van['price'] >= lower_price)]
```

```python
for i in range (101):
    data_van = data_van.loc[(data_van['price'] != i)]
```

```python
data_van['type'] = label.fit_transform(data_van['type'])
```

```python
from sklearn.model_selection import train_test_split

target = data_van['price'].values
pred = data_van.drop(columns=['price'], axis=1).values

# define data training and testing from both attributes
X_train_harga, X_test_harga, y_train_harga, y_test_harga = train_test_split(pred,target, train_size = 0.7, random_state=2022)
```

```python
from sklearn.linear_model import LinearRegression

clf = LinearRegression()
clf.fit(X_train_harga, y_train_harga)
```

```python
y_pred = clf.predict(X_test_harga)
```

```python
data_eval = pd.DataFrame({'Harga (predicted)': y_pred.flatten(), 'Harga (actual)': y_test_harga.flatten() })
data_eval
```

```python
# Visualisasi
data_only= data_eval[['Harga (actual)', 'Harga (predicted)']]       # Mengambil atribut actual dan predicted saja dari dataframe
df_temp = data_only.head(20)
df_temp.plot(kind='bar',figsize=(20,15))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.show()
```

```python
# Nilai MSE
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

MSE = mean_squared_error(y_test_harga,y_pred)
MAE = mean_absolute_error(y_test_harga,y_pred)
RMSE = np.sqrt(MSE)
R2 = r2_score(y_test_harga, y_pred)

print("Nilai MSE: ", MSE)
print("Nilai MAE: ", MAE)
print("Nilai RMSE: ", RMSE)
print("Nilai R2: ", R2)
```

```python
# Preprocessing menggunakan LabelEncoder
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
clean_data['region'] = label.fit_transform(clean_data['region'])
clean_data['manufacturer'] = label.fit_transform(clean_data['manufacturer'])
clean_data['condition'] = label.fit_transform(clean_data['condition'])
clean_data['fuel'] = label.fit_transform(clean_data['fuel'])
clean_data['title_status'] = label.fit_transform(clean_data['title_status'])
clean_data['transmission'] = label.fit_transform(clean_data['transmission'])
clean_data['type'] = label.fit_transform(clean_data['type'])
clean_data['state'] = label.fit_transform(clean_data['state'])
```

```python
bins = pd.qcut(clean_data['price'], q=4)
clean_data['price_bin'] = bins.cat.codes
clean_data.head()
print(bins)
```

```python
clean_data['price'].describe()
```

```python
gr = clean_data.groupby('price_bin').head(10)
gr.head()
```

```python
clean_data.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# X = vehicles.drop(['price'], axis=1)
target = clean_data['price_bin']
pred = clean_data.drop(columns=['price','price_bin'])

# define data training and testing from both attributes
X_train, X_test, y_train, y_test = train_test_split(pred,target, train_size = 0.7, random_state=123)
```

```python
from sklearn.preprocessing import MinMaxScaler

#define the scaler
scaler = MinMaxScaler()

x_train_transformed = scaler.fit_transform(X_train)
x_test_transformed = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# Melakukan training pada model decision tree
clf = DecisionTreeClassifier(random_state=0, criterion='entropy')
clf.fit(x_train_transformed, y_train)
```

```python
# Memprediksi data testing
predicted = clf.predict(x_test_transformed)
```

```python
from sklearn.metrics import precision_score, \
    recall_score, classification_report, \
    accuracy_score, f1_score, ConfusionMatrixDisplay

def evaluate_classifier_performance(classifier, prediction, y_test = y_test):
    # Informasi evaluasi secara compact
    print("Hasil Evaluasi:\n\n%s\n" % (classification_report(y_test, classifier.predict(X_test))))

    # Butuh informasi lebih lengkap? silakan disimak di bawah ini
    print('Accuracy:', accuracy_score(y_test, prediction))
    print('Precision Macro Average:', precision_score(y_test, prediction, average='macro'))
    print('Precision Micro Average:', precision_score(y_test, prediction, average='micro'))
    print('Recall Macro Average:', recall_score(y_test, prediction, average='macro'))
    print('Recall Micro Average:', recall_score(y_test, prediction, average='micro'))
    print('F1 Macro Average:', f1_score(y_test, prediction, average='macro'))
    print('F1 Micro Average:', f1_score(y_test, prediction, average='micro'))

    # Confusion Matrix
    print("\n\nConfusion Matrix\n")
    ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test)
```

```python
# Memanggil fungsi evaluasi 
evaluate_classifier_performance(clf, predicted)
```
